```python
import matplotlib.pyplot as plt
```

# <a id="Bar">● Bar</a>

```python
"""
matplotlib.pyplot.bar(left, height, width=0.8, bottom=None, hold=None, data=None, **kwargs)
https://plot.ly/matplotlib/bar-charts/
"""

label = ['Baseline', 'POS bigram', 'Context words']
y = [25, 50, 75]
width = 1/1.5
plt.ylim(0, 100)
plt.bar(xrange(1, len(label)+1), y, width, color="blue", align='center') # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
plt.xticks(xrange(1, len(label)+1), label) # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xticks
plt.ylabel("Accuracy %", fontsize=16)
plt.show()
```

# <a id="Plot">● Plot</a>

```python
width = [1, 10]
res_list = [30, 80]
plt.ylabel("Accuracy", fontsize=16) # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.ylabel
plt.xlabel("Width", fontsize=16) # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xlabel
plt.title("Accuracy curve", fontsize=16)
plt.ylim(0, 100)
plt.plot(width, res_list)
plt.show()
```
