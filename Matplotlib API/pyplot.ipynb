# <a id="Bar">● Bar</a>

```python
"""
matplotlib.pyplot.bar(left, height, width=0.8, bottom=None, hold=None, data=None, **kwargs)
https://plot.ly/matplotlib/bar-charts/
"""

label = ['Baseline', 'POS bigram', 'Context words']
y = [25, 50, 75]
width = 1/1.5
plt.ylim(0, 100)
plt.bar(range(1, len(label)+1), y, width, color="blue", align='center') # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
plt.xticks(range(1, len(label)+1), label) # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xticks
plt.ylabel("Accuracy %", fontsize=16)
plt.show()
```

```python
# Create lists for the plot
materials = ['PE EDN', 'VI EDN', 'PE EDN Restored', 'VI EDN Restored']

x_pos = np.arange(len(materials))

means = [0.642361, 0.704861, 0.652778, 0.677083]
stds = [0.150312, 0.057265, 0.132946, 0.113154]
#title = 'Shake haptic to Push haptic Category Recognition'
title = 'Shake haptic to Grasp audio Category Recognition'

# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, means, yerr=stds, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylim(0, 1)
ax.set_ylabel('Recognition Accuracy')
ax.set_xticks(x_pos)
ax.set_xticklabels(materials)
ax.set_title(title)
ax.yaxis.grid(True)

file_name = title.replace(" ", "_")+'.png'

# Save the figure and show
plt.tight_layout()
#plt.savefig(file_name, bbox_inches='tight', dpi=100)
plt.show()
```

```python
ind = np.arange(2)
GeneratedAccuracy = np.array([40, 50])
GeneratedAccuracySD = np.array([5, 10])
ActualAccuracy = np.array([45, 60])
ActualAccuracySD = np.array([2, 3])

x_labels = ["A", "B"]

bar_width = 0.40

#plt.subplots(figsize=(10, 5))

plt.bar(ind, GeneratedAccuracy, width=bar_width, yerr=GeneratedAccuracySD, capsize=10, color='blue', label='Projected Features')
plt.bar(ind + bar_width, ActualAccuracy, width=bar_width, yerr=ActualAccuracySD, capsize=10, color='red', label='Ground Truth Features')

plt.yticks(fontsize=14)
plt.xticks(ind+(bar_width/2), x_labels, fontsize=14)
#plt.ylim(0, 100)
title = "Title"
plt.title(title, fontsize=20)
plt.ylabel('% Recognition Accuracy', fontsize=16)

plt.legend(fontsize=16)
plt.show()
```

# <a id="Errorbar">● Errorbar</a>

```python

all_obj_real_gen_acc = [0.46624999999999994, 0.62, 0.6825, 0.645]
all_obj_real_gen_acc_std = [0.046249999999999986, 0.00666666666666671, 0.01749999999999996, 0.015000000000000013]
all_obj_real = [0.37875, 0.445, 0.49750000000000005, 0.54]
all_obj_real_std = [0.03125, 0.018333333333333313, 0.0325, 0.04999999999999999]

plt.figure(figsize=(5, 5))
plt.errorbar(range(1, 5), all_obj_real_gen_acc, yerr=all_obj_real_gen_acc_std, fmt='-o', label="Real + Projected Features Accuracy", color='pink')
plt.errorbar(range(1, 5), all_obj_real, yerr=all_obj_real_std, fmt='-o', label="Real Features Accuracy", color='blue')
plt.xticks(range(1, 5))
plt.ylim(0, 1)
plt.legend()
plt.ylabel("% Recognition Accuracy")
plt.xlabel("Number of Shared Objects Categories")
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
s = np.random.rand(*x.shape) * 800 + 500

#plt.scatter(x, y, s, c="g", alpha=0.5, marker=r'$\clubsuit$', label="Luck")
plt.scatter(x, y, s, c="g", alpha=0.5, marker=r'$\clubsuit$', edgecolor='black', label="Luck")

plt.xlabel("Leprechauns")
plt.ylabel("Gold")
plt.legend(loc='upper left')
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(10)
y = rng.randn(10)
colors = rng.rand(10)
sizes = 1000 * rng.rand(10)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, marker='o', label='Source', edgecolor='black', cmap='viridis')
plt.scatter(y, x, c=colors, s=sizes, alpha=0.3, marker='*', label='Target', edgecolor='black', cmap='viridis')
#ax.scatter(Z1_train_test_obj[:, 0], Z1_train_test_obj[:, 1], c=Y1_train_test_obj[:, 0], marker='o', label='Source: '+source_robot_list[0].capitalize())

plt.legend(loc='center left', bbox_to_anchor=(1.2, 0.5))
plt.colorbar()  # show color scale
plt.show()
```

# <a id="Subplot">● Subplot</a>

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html

```python
fig, ((a,b),(c,d)) = plt.subplots(2,2, figsize=(10, 10))
x = np.arange(10)
a.plot(x, x)
b.plot(x, x)
c.plot(x, x)
d.plot(x, x)

a.set_ylim([0, 10])
a.set_title("A")
a.legend(['video'], loc='upper left')

fig.suptitle('Accuracy Curve', fontsize='20')
fig.subplots_adjust(hspace=0.1)
c.annotate('Accuracy', (0, 0.5), xytext=(-30, 150),
                    textcoords='offset points', xycoords='axes fraction',
                    ha='right', va='center', size=20, rotation=90)
d.annotate('Seconds', (0.5, 1), xytext=(-150, -300),
                    textcoords='offset points', xycoords='axes fraction',
                    ha='center', va='bottom', size=20)
fig.savefig("image3.png", bbox_inches='tight', dpi=100)
```
