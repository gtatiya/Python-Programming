```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
"""
matplotlib.image.imread(fname, format=None)
Read an image from a file into an array.
"""
image = mpimg.imread("image.jpg")
plt.imshow(image, cmap=plt.cm.gray)
plt.plot([50], [50], 'o', c='r')
plt.show()
print(image.shape)
```

```python
def print_image(image, title):
    """Print the image

    :param image: image pixels in list
    :param title: title as string to be printed on top of the image
    """
    #plt.imshow(image)
    #plt.imshow(image, cmap=plt.cm.gray)
    plt.imshow(image, cmap="jet")
    plt.title(title)
    plt.colorbar()
    plt.savefig("image2.jpg", bbox_inches='tight', dpi=100) #remove white space and save image
    plt.show()
    
print_image(image, "Original Image")
```

```python
x_values = 10
y_values = 9

my_data = np.random.random((x_values, y_values))

plt.imshow(my_data.T, vmin=0, vmax=1)

plt.title("Position", fontsize=16)
plt.xlabel("Temporal Bins", fontsize=16)
plt.ylabel("Joints", fontsize=16) 

ax = plt.gca()
ax.set_xticks(np.arange(0, x_values, 1))
ax.set_yticks(np.arange(0, y_values, 1))
ax.set_xticklabels(np.arange(1, x_values+1, 1))
ax.set_yticklabels(np.arange(1, y_values+1, 1))

plt.colorbar()
#plt.colorbar(fraction=0.046)

plt.show()
```
