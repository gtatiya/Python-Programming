# Countdown

```python
test_cases = int(input().strip())

for t in range(test_cases):
    N, K = [int(x) for x in input().strip().split(' ')]
    A_list = [int(x) for x in input().strip().split(' ')]
    K_list = list(range(K, 0, -1))
    
    count = 0
    for i in range(0, N-K+1):
        if A_list[i:i+K] == K_list:
            count += 1
    
    print("Case #"+str(t+1)+": "+ str(count))
```

# Perfect Subarray

```python
import math

test_cases = int(input().strip())

for t in range(test_cases):
    N = [int(x) for x in input().strip().split(' ')]
    A_list = [int(x) for x in input().strip().split(' ')]
    
    count = 0
    for i in range(len(A_list)):
        for j in range(i, len(A_list)):
            if math.sqrt(sum(A_list[i:j+1])) % 1 == 0:
                count += 1

    print("Case #"+str(t+1)+": "+ str(count))
```

# Candies

```python
test_cases = int(input().strip())

for t in range(test_cases):
    N, Q = [int(x) for x in input().strip().split(' ')]
    A_list = [int(x) for x in input().strip().split(' ')]
    
    query_sum = 0
    for _ in range(Q):
        query = [x for x in input().strip().split(' ')]
        s, e = int(query[1])-1, int(query[2])
        
        if query[0] == "Q":
            length = e-s
            for i in range(length):
                if i == 1:
                    query_sum -= A_list[s+i] * (i+1)
                else:
                    query_sum += A_list[s+i] * (i+1)
        elif query[0] == "U":
            A_list[s] = e
    
    print("Case #"+str(t+1)+": "+ str(query_sum))
```
