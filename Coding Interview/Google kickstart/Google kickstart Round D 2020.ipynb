# Record Breaker

```python
test_cases = int(input().strip())

for t in range(test_cases):
    N = int(input().strip())
    V_list = [int(x) for x in input().strip().split(' ')]
    
    num_rb = 0
    for i in range(N):
        if i+1 < N and V_list[i] > V_list[i+1] or i+1 == N:
            if len(V_list[:i]) > 0 and max(V_list[:i]) < V_list[i] or len(V_list[:i]) == 0:
                num_rb += 1                
    
    print("Case #"+str(t+1)+": "+ str(num_rb))
```

# Alien Piano

```python
test_cases = int(input().strip())

pitch = ['A', 'B', 'C', 'D']

for t in range(test_cases):
    N = int(input().strip())
    V_list = [int(x) for x in input().strip().split(' ')]
    
    rule_break = 0
    pitch_idx = 0
    pitch_song = pitch[pitch_idx]
    
    for i in range(N-1):
        if V_list[i+1] > V_list[i] and pitch_idx < 4-1:
            pitch_idx += 1
            pitch_song += pitch[pitch_idx]
        elif V_list[i+1] < V_list[i] and pitch_idx > 0:
            pitch_idx -= 1
            pitch_song += pitch[pitch_idx]
        elif V_list[i+1] == V_list[i]:
            pitch_song += pitch[pitch_idx]
        else:
            rule_break += 1
            pitch_idx = 0
            pitch_song += pitch[pitch_idx]
        
#         print("pitch_song: ", pitch_song)
    
    print("Case #"+str(t+1)+": "+ str(rule_break))
```
