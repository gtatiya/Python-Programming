# Bike Tour

```python
test_cases = int(input().strip())

for i in range(test_cases):
    N = int(input().strip())
    H_list = [int(x) for x in input().strip().split(' ')]
    
    count = 0
    for k in range(N):
        if (k > 0) and (k < N-1):
            before_h = H_list[k-1]
            after_h = H_list[k+1]
            if (before_h < H_list[k]) and (H_list[k] > after_h):
                count += 1
    
    print("Case #"+str(i+1)+": "+ str(count))
```

# Bus Routes

```python
test_cases = int(input().strip())

for i in range(test_cases):
    N, D = [int(x) for x in input().strip().split(' ')]
    X_list = [int(x) for x in input().strip().split(' ')]
    
    latest_day = D
    for j in range(N-1, -1, -1):
        latest_day = int(latest_day/X_list[j])*X_list[j]

    print("Case #"+str(i+1)+": "+ str(latest_day))
```

# Robot Path Decoding

```python
test_cases = int(input().strip())

for i in range(test_cases):
    program_str = input().strip()
    
    while '(' in program_str:
        first_i = program_str.rfind('(')
        
        program_str_temp = program_str
        last_i = program_str_temp.find(')')

        while last_i < first_i:
            original_len = len(program_str_temp)
            program_str_temp = program_str_temp[first_i:]
            diff_len = original_len - len(program_str_temp)
            
            last_i = program_str_temp.find(')')
            last_i = diff_len+last_i
            
        x = int(program_str[first_i-1])

        temp = ""
        for _ in range(x):
            temp += program_str[first_i+1:last_i]

        program_str = program_str[0:first_i-1] + temp + program_str[last_i+1:]
    
    x, y = 1, 1
    for move in program_str:
        if move == 'N':
            x -= 1
            if x == 0:
                x = 10 ** 9
        elif move == 'S':
            x += 1
            if x == 10 ** 9 + 1:
                x = 1
        elif move == 'W':
            y -= 1
            if y == 0:
                y = 10 ** 9
        elif move == 'E':
            y += 1
            if y == 10 ** 9 + 1:
                y = 1
    
    print("Case #"+str(i+1)+": "+ str(y) + " " + str(x))
    
```
