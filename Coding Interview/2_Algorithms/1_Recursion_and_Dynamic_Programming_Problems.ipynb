# Recursion



Input:
88

Output:
[tt, tu, tv, ut, uu, uv, vt, vu, vv]

```python
num_char = {"1": [""], "2": ["A", "B", "C"], "3": ["D", "E", "F"], "4": ["G", "H", "I"], "5": ["J", "K", "L"],
            "6": ["M", "N", "O"], "7": ["P", "Q", "R", "S"], "8": ["T", "U", "V"], "9": ["W", "X", "Y", "Z"], "0": [" "]}

number = "88" # [tt, tu, tv, ut, uu, uv, vt, vu, vv]

# Brute force
for first_char in num_char[number[0]]:
    for second_char in num_char[number[1]]:
        print(first_char+second_char)
```

```python
def get_combinations(number):
    
    first_chars = num_char[number[0]]
    
    # Base case
    if len(number) == 1:
        #print("results: ", number, first_chars)
        return first_chars
    
    # Recurrsion
    results = []
    for combindation in get_combinations(number[1:]):
        for char1 in first_chars:
            results.append(char1+combindation)
    
    #print("results: ", number, results)
    return results


number = "204" # 88 --> [tt, tu, tv, ut, uu, uv, vt, vu, vv]
print(get_combinations(number))
```

# Q) [The Coin Change Problem](https://www.hackerrank.com/challenges/coin-change)

```python
def getWays(amount, denoms):
    
    # dp[j] stores the number of solutions for j
    # For base case j=0, number of solutions is 1 (not using any coin)
    dp = [1]+[0]*amount
    # print("dp: ", dp)
    for coin in denoms:
        # print('coin: ', coin)
        for j in range(coin, amount+1):
            # print('j: ', j)
            dp[j] += dp[j-coin]
            # print("dp: ", dp)

    return dp[-1]

amount = 8 #20
denoms = [3, 2, 1] #[25, 10, 5, 1]
print(getWays(amount, denoms))
```

# Q) [Stock Maximize](https://www.hackerrank.com/challenges/stockmax)

```python
def stockmax(prices):
    
    max_price = prices.pop()
    
    max_price_sum = 0  
    price_sum = 0
    
    # Iterating reversed because we have to buy before selling
    for current_price in reversed(prices):
        max_price = max(max_price, current_price)
        max_price_sum += max_price # Sell for better price
        price_sum += current_price # Buy on each day, max_price_sum will be more than price_sum for better price and same for no transaction
        
    return max_price_sum - price_sum

# prices = [1, 2, 3, 4] # 6
# prices = [5, 4, 3, 4, 5] # 4
# prices = [5, 3, 2] # 0
# prices = [1, 2, 100] # 197
prices = [1, 3, 1, 2] # 3

print(stockmax(prices))
```
