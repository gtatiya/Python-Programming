```python
import matplotlib.pyplot as plt
import numpy as np
```

# Logarithm

```python
x = np.linspace(0.1, 100, num=200)

y = np.log(x)
plt.plot(x, y, label='Natural log')

y = -np.log(x)
plt.plot(x, y, label='-Natural log')

y = np.log2(x)
plt.plot(x, y, label='Base-2 log')

y = np.log10(x)
plt.plot(x, y, label='Base-10 log')

plt.ylabel("Y", fontsize=16)
plt.xlabel("X", fontsize=16)
plt.title("Logarithm", fontsize=16)
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(0.01, 2, num=200)

y = np.log(x)
plt.plot(x, y, label='Natural log')

y = -np.log(x)
plt.plot(x, y, label='-Natural log')

plt.ylabel("Y", fontsize=16)
plt.xlabel("X", fontsize=16)
plt.title("Logarithm", fontsize=16)
plt.legend()
plt.grid()
plt.show()
```

# Exponential

```python
x = np.linspace(-5, 5, num=200)

y = np.exp(x)
plt.plot(x, y, label='Exponential')

y = np.exp2(x)
plt.plot(x, y, label='2**p')

plt.ylabel("Y", fontsize=16)
plt.xlabel("X", fontsize=16)
plt.title("Exponential", fontsize=16)
plt.legend()
plt.grid()
plt.show()
```

# Softmax function (normalized exponential function)

Takes as input a vector z of K real numbers, and normalizes it into a probability distribution consisting of K probabilities proportional to the exponentials of the input numbers.

\begin{align}
{\displaystyle \sigma (\mathbf {z} )_{i}={\frac {e^{z_{i}}}{\sum _{j=1}^{K}e^{z_{j}}}}{\text{ for }}i=1,\dotsc ,K{\text{ and }}\mathbf {z} =(z_{1},\dotsc ,z_{K})\in \mathbb {R} ^{K}}
\end{align}

```python
from scipy.special import softmax
np.set_printoptions(precision=5)

x = np.random.randint(low=0, high=5, size=10)
y = softmax(x)

print("X: ", x)
print("Y: ", y)
print("Sum: ", y.sum())

plt.bar(range(len(y)), y)
plt.ylabel("Probability", fontsize=16)
plt.xlabel("Variable", fontsize=16)
plt.title("Softmax", fontsize=16)

plt.show()
```
